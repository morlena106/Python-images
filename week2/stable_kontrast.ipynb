```python
#стабильный автоконстраст 
#тбрасывает пиксели в мак и мин контрастом 
from skimage.io import imread, imsave, imshow
img = imread('https://stepik.org/media/attachments/lesson/58402/tiger-low-contrast.png')
```

```python
imshow(img)
```

```python
from matplotlib.pyplot import hist
values, bin_edges, patches = hist(img.ravel(), bins=range(257))
```

```python
pix = img.shape[0]*img.shape[1]
k = round(pix*0.05)
imgg = img.copy()
img_n = imgg.ravel()
img_n.sort()
a = img_n[k]
b = img_n[pix - k]
print(a,b)
```

```python
s = int(round((b-a)*0.05))
img_3 = clip(img,a+s,b-s)
```

```python
imshow(img_3)
```

```python
img_1=img_3-img_3.min()
t = img_3.max()-img_3.min()
r = 255/t
img_2 = img_1*r
```

```python
values, bin_edges, patches = hist(img_2.ravel(), bins=range(257))
```

```python
imshow(img_2.astype('uint8'))
```

```python
#верный код 
from skimage.io import imread, imsave, imshow
#img = imread('img.png')
img.astype('float')
imgg = img.copy()
img_n = imgg.ravel()
img_n.sort()
k = round(len(img_n)*0.05)
a = img_n[k]
b = img_n[-k]
from numpy import clip
img_3 = clip(img,a,b)
img_1=img_3-img_3.min()
t = img_3.max()-img_3.min()
r = 255/t
img_2 = img_1*r
imshow(img_2.astype('uint8'))
```
